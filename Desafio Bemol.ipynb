```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
realizado = pd.read_excel('dados.xlsx', sheet_name = 'realizado')
```

```python
realizado
```

```python
orcado = pd.read_excel('dados.xlsx', sheet_name = 'orcado')
```

```python
orcado
```

```python
realizado.T.reset_index(drop = True)[1:]
```

```python
realizado = realizado.T.reset_index(drop = True)[1:]
```

```python
realizado.columns = ['mês', 'realizado']
```

```python
realizado
```

```python
df = pd.merge(orcado, realizado, how = 'inner', on = 'mês')
```

```python
df
```

```python
df['diff'] = df['orcado'] - df['realizado']
df
```

```python
df.to_csv('output.csv')
```

```python
fig, ax = plt.subplots(figsize = (20, 10))

ax.bar(df['mês'], df['orcado'], color = 'blue', label = 'Orçado')
ax.bar(df['mês'], df['realizado'], color = 'darkorange', label = 'Realizado')
ax.set_xlabel('Mes')
ax.set_ylabel('$')
ax.set_title('Gráfico Orçamento')
ax.legend(loc = 2)
```

```python
fig.savefig('Barplot.png')
```
